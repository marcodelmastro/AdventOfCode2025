## Day 2

https://adventofcode.com/2025/day/2

```python
import re 

def validID1(ID):
    return not bool(re.fullmatch(r"(\d+)\1", str(ID))) # two identical halves

def validID2(ID):
    # ^(.*?) --> smallest possible substring
    # \1+ --> rest of the string is made by repetitions of that substring
    return not bool(re.match(r"^(.*?)\1+$", str(ID))) # repetition of same substring

def solve2(filename,part=1):
    f = open(filename)
    IDs = [ [int(i) for i in b.split("-")] for b in f.read().split(",") ]
    sumID = 0
    for i1,i2 in IDs:
        for i in range(i1,i2+1):
            if part==1:
                if not validID1(i):
                    sumID += i
            elif part==2:
                if not validID2(i):
                    sumID += i
    return sumID
```

```python
print("Test 1:",solve2("examples/example02.txt",part=1))
print("Part 1:",solve2("AOC2025inputs/input02.txt",part=1))
print("Test 2:",solve2("examples/example02.txt",part=2))
print("Part 3:",solve2("AOC2025inputs/input02.txt",part=2))
```
