## Day 2

https://adventofcode.com/2025/day/2

```python
import re 

def validID1(ID):
    return not bool(re.fullmatch(r"(\d+)\1", str(ID))) # two identical halves

def validID2(ID):
    return not bool(re.fullmatch(r"(\d+)\1+", str(ID))) # one or more repetitions of same substring

def solve2(filename):
    f = open(filename)
    sum1 = 0
    sum2 = 0
    for i1,i2 in [ [int(i) for i in b.split("-")] for b in f.read().split(",") ]:
        for i in range(i1,i2+1):
            if not validID1(i):
                sum1 += i
            if not validID2(i):
                sum2 += i
    return sum1,sum2
```

```python
t1, t2 = solve2("examples/example02.txt")
s1, s2 = solve2("AOC2025inputs/input02.txt")

print("Test 1:",t1)
print("Part 1:",s1)
print("Test 2:",t2)
print("Part 2:",s2)
```
