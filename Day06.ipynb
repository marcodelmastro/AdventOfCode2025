https://adventofcode.com/2025/day/6

```python
import numpy as np
from math import prod
```

```python
def part1(filename):
    with open(filename) as f:
        worksheet = [l.strip().split() for l in f.readlines()]
        operations = worksheet[-1]
        values = np.array([ [int(i) for i in l] for l in worksheet[:-1]])
        values = np.flipud(np.rot90(values))    
        s = 0
        for v,o in zip(values,operations):
            if o=="*":
                r = int(prod(v))
            elif o=="+":
                r = int(sum(v))
            s += r
        return s
```

```python
print("Test 1:", part1("examples/example06.txt"))
print("Part 1:", part1("AOC2025inputs/input06.txt"))
```

```python
def part2(filename):
    with open(filename) as f:
        worksheet = np.array([[c for c in l.strip("\n")[::-1]] for l in f.readlines()]) # Split all characters
        worksheet = np.flipud(np.rot90(worksheet)) # Rotate worksheet
        s = 0
        values = []
        for w in worksheet:
            v = "".join(c for c in w[:-1]) # Form number
            try: # Try str->int conversion. If it fails we reached the end of values for current operation
                 # Note that this is not true only for last operation: don't put operation in except block!
                values.append(int(v.strip()))
                o = w[-1] # Catch operator. Normally " ", if it's * or + execute operation and reset values array
                if o=="*":
                    r = int(prod(values))
                    s += r
                    values = []
                elif o=="+":
                    r = int(sum(values))
                    s += r
                    values = []
            except:
                continue
    return s 
```

```python
print("Test 2:", part2("examples/example06.txt"))
print("Part 2:", part2("AOC2025inputs/input06.txt"))
```
