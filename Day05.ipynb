## Day 5

https://adventofcode.com/2025/day/5

```python
def read_input_5(filename):
    with open(filename) as f:
        c = f.read().split("\n\n")
        ranges = [[int(i) for i in l.split("-")] for l in c[0].split("\n")]
        values = [int(i) for i in c[1].strip().split('\n')]
        return ranges, values
```

```python
def is_fresh(v,ranges):
    for a,b in ranges:
        if a<= v <=b:
            return True
    return False

def part1(filename):
    ranges, values = read_input_5(filename)
    return sum([is_fresh(v,ranges) for v in values ])
```

```python
print("Test 1:", part1("examples/example05.txt"))
print("Part 1:", part1("AOC2025inputs/input05.txt"))
```

```python
def part2(filename):
    intervals, _ = read_input_5(filename)

    # sort intervals by lowest boundary
    intervals = sorted(intervals, key=lambda x : x[0])
    
    # merge overlapping and contigous intervals
    stack = []
    stack.append(intervals[0])
    for i in range(1,len(intervals)):
        last_interval = stack[-1]
        if last_interval[1] >= intervals[i][0] or last_interval[1] == intervals[i][0]-1:
            last_interval[1] = max(intervals[i][1],last_interval[1])
            stack.pop(len(stack)-1)
            stack.append(last_interval)
        else:  
            stack.append(intervals[i])
     
    # sum spans of all non-overlapping intervals
    return sum([b-a+1 for a,b in stack ])
```

```python
print("Test 2:", part2("examples/example05.txt"))
print("Part 2:", part2("AOC2025inputs/input05.txt"))
```
