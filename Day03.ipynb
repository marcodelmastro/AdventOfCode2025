```python
def largest_joltage_1(b):
    # find largest digit in series, leaving space for another digit after it
    d = max(b[:len(b)-1])
    # find second largest digit in remaining sequence right of first digit
    i = b.index(d)
    u = max(b[i+1:])
    return 10*d+u

def largest_joltage(b, k=12):
    # generalizes part 1 solution to any number of digits
    n = len(b)
    result = []
    start = 0
    for remaining in range(k, 0, -1): # loop on all needed digits
        # Define range ends such that enough digits remain
        end = n - remaining + 1
        # Find the largest digit in b[start:end]
        max_digit = max(b[start:end])
        max_index = b.index(max_digit, start, end)
        # Collect digit
        result.append(max_digit)
        # Redefine beginning of search interval, repeat for following digits
        start = max_index + 1
    return int("".join([str(i) for i in result]))

def part1(filename):
    with open(filename,'r') as f:
        #return sum([largest_joltage_1(b) for b in [ [int(n) for n in l.strip() ] for l in f.readlines() ] ])
        return sum([largest_joltage(b,2) for b in [ [int(n) for n in l.strip() ] for l in f.readlines() ] ])

def part2(filename):
    with open(filename,'r') as f:
        return sum([largest_joltage(b,12) for b in [ [int(n) for n in l.strip() ] for l in f.readlines() ] ])
```

```python
print("Test 1:",part1("examples/example03.txt"))
print("Part 1:",part1("AOC2025inputs/input03.txt"))
print("Test 2:",part2("examples/example03.txt"))
print("Part 2:",part2("AOC2025inputs/input03.txt"))
```
